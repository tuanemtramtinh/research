```python
#import library

import spacy
from textacy import extract, preprocessing
```

```python
#open input
with open("./input.txt", "r", encoding="UTF-8") as f:
  input = f.read()
  
```

```python
#clean input
input = preprocessing.normalize.whitespace(input)

print(input)
```

```python
#initial spacy
nlp = spacy.load("en_core_web_lg")
doc = nlp(input)

print(doc)
```

```python
#extract noun_chunks
chunks = extract.noun_chunks(doc, min_freq=1)

#remove noun which have role is Pronoun
chunks = [chunk for chunk in chunks if chunk.root.pos_ != 'PRON']
```

```python
#extract S-V-O

triplets = list(extract.triples.subject_verb_object_triples(doc))


print (triplets)
# for subj, verb, obj in triplets:
#     print(f"S: {subj} | V: {verb} | O: {obj}")
```
